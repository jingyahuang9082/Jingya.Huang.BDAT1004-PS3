```python
#Step 1 import libarary
import pandas as pd
#Step 2 and Step 3 import data and called it 'users'
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

```python
users.columns
```

```python
#Step 4 the mean age per occupation
users.groupby('occupation').mean(['age'])
```

```python
#Step 4 continus : the mean age per occupation
users.groupby('occupation').agg({'age':'mean'})
```

```python
#Step 5 Discover the Male ratio per occupation and sort it from the most to the least
gender_count_by_occupation= users.groupby(['occupation', 'gender']).gender.count()
total_people = users.groupby('gender').gender.count()
print((gender_count_by_occupation /total_people).sort_values(ascending=False).xs('M',level=1))
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages
users.groupby('occupation').agg({'age':['min','max']})

```

```python
#Step 7 calculate the mean age by occupation and sex
users.groupby(['occupation','gender']).agg({'age':'mean'})
```

```python
# Step 8 For each occupation present the percentage of women and men
total_People_By_Occupation = users.groupby('occupation').gender.count()
total_Gender_By_Occupation = users.groupby(['occupation', 'gender']).gender.count()
print(total_Gender_By_Occupation*100/total_People_By_Occupation)
```

The first three columns are year, month, and day. The remaining 12 columns are

```python
# Step 1 import libraries  
import numpy as np
import pandas as pd
import datetime

#Step 2 and Step 3
data=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data',sep="\s+",
                 parse_dates={'datetime':['Yr','Mo','Dy']})
data
```

```python
# Step 4 Year 2061? Do we really have data from this year? Create a function to fix it and apply it
def fix_century(x):
    year=x.year-100 if x.year > 1979 else x.year
    
    return datetime.date(year,x.month,x.day)

data['datetime']=data['datetime'].apply(fix_century)
data
```

```python
# Step 5.Set the right dates as the index.
data.set_index(["datetime"],inplace=True)
data
```

```python
#Step 6. missing values there are in total.
data.isna().sum()
```

```python
# Step 7 non-missing values there are in total.
data.notna().sum()
```

```python
# Step 8 Calculate the mean windspeeds of the windspeeds over all the locations and all the times
data.mean().mean()
```

```python
# Step 9 Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats=pd.DataFrame()
loc_stats['min']=data.min(axis=0)
loc_stats['max']=data.max(axis=0)
loc_stats['mean']=data.mean(axis=0)
loc_stats['standardDeviations ']=data.std(axis=0)
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats=pd.DataFrame()
day_stats['min']=data.min(axis=1)
day_stats['max']=data.max(axis=1)
day_stats['mean']=data.mean(axis=1)
day_stats['stdstandardDeviations']=data.std(axis=1)
day_stats
```

```python
# Step 11 Find the average windspeed in January for each location
data_index = data['date'].apply(lambda x:x.month)
january_average=data.query('month==1')
january_average.loc[:,'RPT':'MAL'].mean()
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
data.asfreq('Y')
```

```python
# Step 13 Downsample the record to a monthly frequency for each location.
data.asfreq('M')
```

```python
# Step 14 Downsample the record to a weekly frequency for each location.
data.asfreq('W')
```

```python
# Step 15 Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks
df = data[data.index < pd.to_datetime('1962-01-01')]
df.asfreq('W').mean()
df.asfreq('W').min()
df.asfreq('W').max()
df.asfreq('W').std()
```

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were 
assassinated.


```python
import matplotlib.pyplot as plt
import pandas as pd

roman= pd.read_csv('C:\\Users\\jane\\Downloads\\roman-emperor-reigns.csv')
assassinated_death = roman[roman['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]
emperor = assassinated_death["Emperor"]
cause_of_death = assassinated_death["Cause_of_Death"]
plt.pie(range(len(cause_of_death)), labels=emperor,autopct='%1.2f%%', startangle=40, radius=0.045 * 100,rotatelabels = 300)
fig = plt.figure(figsize=[15, 10])
plt.title("Roman Emperor Reigns")
plt.show()
```
